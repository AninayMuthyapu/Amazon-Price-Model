```python
SYSTEM_PROMPT = """Create a concise description of a product. Respond only in this format. Do not include part numbers.
Title: Rewritten short precise title
Category: eg Electronics
Brand: Brand name
Description: 1 sentence description
Details: 1 sentence on features"""
print(items[0].full)
```

```python
messages = [{"role": "system", "content": SYSTEM_PROMPT}, {"role": "user", "content": items[0].full}]
response = completion(messages=messages, model="groq/openai/gpt-oss-20b", reasoning_effort="low")

print(response.choices[0].message.content)
print()
print(f"Input tokens: {response.usage.prompt_tokens}")
print(f"Output tokens: {response.usage.completion_tokens}")
print(f"Cost: {response._hidden_params['response_cost']*100:.3f} cents")
```

```python
MODEL = "openai/gpt-oss-20b"
```

```python
def make_jsonl(item):
    body = {"model": MODEL, "messages": [{"role": "system", "content": SYSTEM_PROMPT}, {"role": "user", "content": item.full}], "reasoning_effort": "low"}
    line = {"custom_id": str(item.id), "method": "POST", "url": "/v1/chat/completions", "body": body}
    return json.dumps(line)
```

```python
def make_file(start, end, filename):
    batch_file = filename
    with open(batch_file, "w", encoding="utf-8") as f:
        for i in range(start, end):
            f.write(make_jsonl(items[i]))
            f.write("\n")
```

```python
make_file(0, 1000, "jsonl/0_1000.jsonl")
```

```python
import os
from groq import Groq

groq = Groq(api_key=os.environ.get("GROQ_API_KEY"))
```

```python
with open("jsonl/0_1000.jsonl", "rb", encoding="utf-8") as f:
    response = groq.files.create(file=f, purpose="batch")
response
```

```python
file_id = response.id
file_id
```

```python
response = groq.batches.create(completion_window="24h", endpoint="/v1/chat/completions", input_file_id=file_id)
response
```

```python
result = groq.batches.retrieve(response.id)
result
```

```python
response = groq.files.content(result.output_file_id)
response.write_to_file("jsonl/batch_results.jsonl")
```

```python
with open("jsonl/batch_results.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        json_line = json.loads(line)
        id = int(json_line["custom_id"])
        summary = json_line["response"]["body"]["choices"][0]["message"]["content"]
        items[id].summary = summary
```

### Summary of the Notebook

The primary goal of `data_preprossing.ipynb` is to **clean and standardize a messy dataset using AI**. 

It downloads a dataset of raw, unstructured Amazon product descriptions, feeds them into a Large Language Model (LLM), and instructs the LLM to rewrite each one into a clean, structured format (Title, Category, Brand, Description, Details). Once all the items have been summarized by the LLM, the notebook splits the cleaned data into training, validation, and testing sets, and uploads it back to Hugging Face so it can be used to train models later.

***

### Why are we dividing the data into batches?

The notebook divides the items into files of 1,000 requests and uses the **Batch API** (specifically Groq's Batch API) instead of sending requests one by one in a standard `for` loop. There are three major reasons for this:

1. **Bypassing Rate Limits:** API providers strictly limit how many requests you can make per minute (e.g., 500 requests/minute). If you try to run 20,000 (or 800,000) synchronous requests through a loop, the API will block you almost instantly, and your code will crash. The Batch API accepts massive files upfront and processes them at its own pace without rate-limiting you.
2. **Significantly Lower Cost:** Because batch processing is asynchronous (meaning you don't need an answer *right now*), API providers run these jobs when they have idle server capacity. To incentivize this, Batch APIs are usually **50% cheaper** than standard real-time API calls.
3. **Connection Stability:** Making hundreds of thousands of individual HTTP requests over the internet is prone to connection timeouts and network errors. Uploading a single `.jsonl` file with 1,000 requests is much more stable and guarantees all items are received by the server.
